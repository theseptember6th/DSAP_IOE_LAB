```python
import numpy as np
```

# TASK 1
## Draw graphs of the function
- ## `y1=cos(pi*t)`
- ## `y2=sin(2*pi*t)`
- ## `z=y1+y2` where `0<=t<=5`

```python
pi=np.pi
t=np.arange(0,5,0.01)
y1=np.cos(pi*t)
y2=np.sin(2*pi*t)

plt.plot(t,y1,t,y2)
plt.legend(["y1=cos(pi*t)","y2=sin(2*pi*t)"],loc="upper right")
plt.xlabel("t")
plt.ylabel("y")
plt.title("Sinusoidal graphs")
plt.grid()
plt.show()
```

```python
Z=y1+y2
plt.plot(t,Z)
plt.xlabel("t")
plt.ylabel("Z")
plt.grid()
plt.title("sum of sinusoidals")
plt.legend(["Z=y1+y2"],loc="upper right")
plt.show()
```

# **TASK 2** 
## Draw the graph of the function
- ## `Y=e^(n*x)` on interval `-1<=x<=1` for `n=1,2,..4.`
- ## Use subplots

```python
x=np.arange(-1,1,0.01)

for n in range(1,5,1):
    Y=np.exp(n*x)
    plt.subplot(2,2,n)
    plt.grid()
    plt.xlabel(f"for n ={n}")
    plt.ylabel("Y")
    plt.plot(x,Y)
    plt.legend([f"Y=np.exp({n}*x)"],loc="upper right")

plt.tight_layout()
plt.show()



```
